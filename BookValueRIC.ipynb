```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
data = pd.read_csv('RawData/FirmData.csv')
data["Year"] = pd.to_datetime(data["Date"], dayfirst=True).dt.year
data.head(5)
```

```python
df_Book_Value = data.pivot(index="RIC", columns="Year", values="Total Equity")
for i in df_Book_Value.columns:
    df_Book_Value[i] = pd.to_numeric(df_Book_Value[i], errors='coerce')
df_Book_Value.head(5)
```

```python
df_Book_Value = df_Book_Value[~(df_Book_Value[df_Book_Value.columns] == 0)]
```
