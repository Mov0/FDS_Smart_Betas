```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
# Read data
data = pd.read_csv('RawData/FirmData.csv')
data["Year"] = pd.to_datetime(data["Date"], dayfirst=True).dt.year
data.head(5)
```

```python
# Convert dataframe to correct format
df_Market_Cap = data.pivot(index="RIC", columns="Year", values="Company Market Cap")
df_Market_Cap.head(5)
```

```python
# Exclude zeros
df_Market_Cap = df_Market_Cap[~(df_Market_Cap[df_Market_Cap.columns] == 0)]
```

```python
# Check if all columns have data type float64 e.g. only contain numbers
df_Market_Cap.dtypes
```

```python
# Save dataframe
df_Market_Cap.to_pickle("UsableData/MarketCap.pkl")
```
